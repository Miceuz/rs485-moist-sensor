```python
d = loadtxt('./initial-value-distribution-cleanup.csv', delimiter=',')
plot(d)
m = d[:,0]
figure()
plot(m)
```

```python
last = 0
r=[]
for v in m:
    if last != v:
        r.append(v)
        last=v

plot(r)
```

```python
print ("Mean: " + str(mean(r)))
print ("Median: " + str(median(r)))
stddev=sqrt(sum((r-median(r))**2)/size(r))
print ("Standard deviation: " + str(stddev))
hist(r)
xlabel("Value in free air")
ylabel("Count")
```
